# Expectations and Math #
### SCs Only ###
*Note: Same as Branching Process 3 Outcomes Selection*
* Neutral Branching Process
    * $P(n > 0|t) = \frac{1}{1+t}$
    * If alive, size $\approx t + 1$
    * $\rho(n|n > 0, t) = e^{-\frac{n}{t+1}}$
* Selection
    * $\langle{n}\rangle = 1 + \frac{2}{3}s$
    * $\frac{dn}{ds} = ns$
    * $\Longrightarrow n(t) = n(0)e^{st}$
    * exponential growth takes effect at $t \approx \frac{1}{s} \Longrightarrow n \approx \frac{1}{s}$
    * $P(n>0|t=s) \approx s$
    * if $ns<1$, drift dominates
    * if $ns>1$ selection dominates

### SCs, Ps, and Ds ###
* From http://www.ncbi.nlm.nih.gov/pmc/articles/PMC3099568/
    * ? $s_{sc} = 0.08$ ?
    * $K \approx 11,000 \Longrightarrow T_{d} = \log_{2}{K} \approx 14$
    * $r_{s} \approx \frac{1}{40}$ divisions per wk.
    * $r_{p} = \lambda = \frac{1}{6.7}$ divisions per wk.
    * $r_{d} = \mu = \frac{1}{6}$ divisions per wk.
* From theoretical derivation with Non-stochastic SCs for homeostasis ($N_s=N_p=N_d$)
    * $r_p = -\frac{3r_s}{2s_p}$
    * $r_d = r_s + r_p$
    * $N_p(t) = \int\limits_{1}^{t} r_s(N_s(x) - N_p(x))\cdot dx \approx \sum\limits_{n=1}^{t} r_s(N_s(n) - N_p(n))$
        * $\Longrightarrow \frac{d}{dt}N_p(t) = r_s(N_s(t) - N_p(t))$
        * Note: $N_s(t)=N_s$ is a constant for purposes of calculation of $N_p(t)$ and $N_d(t)$ 
* From theoretical derivation with Non-Stochastic SCs for Homeostasis $N_s \neq N_p \neq N_d$
    * $\frac{d}{dt}N_p(t) = N_sr_s + N_pr_p \big(\frac{2s_p}{3}\big) = 0$
        * $\Longrightarrow N_p = -\frac{3N_sr_s}{2r_ps_p}$
        * $\Longrightarrow r_p = -\frac{3N_sr_s}{2N_ps_p}$
    * $\frac{d}{dt}N_d(t) = N_pr_p \big(1 - \frac{2}{3}s_p \big) - N_dr_d$
        * $\Longrightarrow N_d = \big(-\frac{3N_sr_s}{2r_ds_p}\big) \big(1 - \frac{2}{3}s_p \big)$
        * $\Longrightarrow r_d = \big(-\frac{3N_sr_s}{2N_ds_p}\big) \big(1 - \frac{2}{3}s_p \big)$

### SCs, Ps, and Ds (from Study)###
* ? $s_{sc} = 0.08$ ?
* $K \approx 11,000 \Longrightarrow T_{d} = \log_{2}{K} \approx 14$
* $r_{s} \approx \frac{1}{40}$ divisions per wk.
* $r_{p} = \lambda = \frac{1}{6.7}$ divisions per wk.
* $r_{d} = \mu = \frac{1}{6}$ deaths per wk.

### Non-Stochastic SCs for Homeostasis $N_s = N_p = N_d$###
* $r_p = -\frac{3r_s}{2s_p}$
* $r_d = r_s + r_p$
* $N_p(t) = \int\limits_{1}^{t} r_s(N_s(x) - N_p(x))\cdot dx \approx \sum\limits_{n=1}^{t} r_s(N_s(n) - N_p(n))$
    * $\Longrightarrow \frac{d}{dt}N_p(t) = r_s(N_s(t) - N_p(t))$
    * Note: $N_s(t)=N_s$ is a constant for purposes of calculation of $N_p(t)$ and $N_d(t)$ 
    
### Non-Stochastic SCs for Homeostasis $N_s \neq N_p \neq N_d$###
* $\frac{d}{dt}N_p(t) = N_sr_s + N_pr_p \big(\frac{2s_p}{3}\big) = 0$
    * $\Longrightarrow N_p = -\frac{3N_sr_s}{2r_ps_p}$
    * $\Longrightarrow r_p = -\frac{3N_sr_s}{2N_ps_p}$
* $\frac{d}{dt}N_d(t) = N_pr_p \big(1 - \frac{2}{3}s_p \big) - N_dr_d$
    * $\Longrightarrow N_d = \big(-\frac{3N_sr_s}{2r_ds_p}\big) \big(1 - \frac{2}{3}s_p \big)$
    * $\Longrightarrow r_d = \big(-\frac{3N_sr_s}{2N_ds_p}\big) \big(1 - \frac{2}{3}s_p \big)$

# 1st Incomplete Moment Functions #

```python
from scipy import integrate

def uniqueFunc(a): #function that returns the unique values of a
    a = np.sort(a)
    b = np.diff(a)
    b = np.r_[1, b]
    return a[b != 0]

def firstIncompleteMoment(NpValues): #NpValues is a 2-D ndarray of all values of Np in time, it has I rows and T cols, dtype='float64'
    NpValues = NpValues.flatten() #turn it all into one 1-D array
    unique, counts = np.unique(NpValues, return_counts=True)
    rho = (dict(zip(unique, counts/NpValues.size))) #dictionary of values of rho for each n, found by the probability of n appearing in NpValues
#     print(rho)
    #sort in place so that each row is in numerical order 
    NpValues.sort()
#     print(NpValues)
    #make array of x values
    mValues = uniqueFunc(NpValues)
    #take out non-unique values
#     for row in range(0, NpValues.shape[0]):
#         mValues[row, 0:((uniqueFunc(NpValues[row])).shape[0])] = uniqueFunc(NpValues[row])
#     mValues.sort(axis=1)
#     print(mValues)
    nRhoValues = mValues.copy() #copy the set of cVales into mValues, so that nRhoValues can be manipulated in place
    #multiply each n value in NpValues by corresponding rho
    for n in np.nditer(nRhoValues, op_flags=['readwrite']):
        if(n != 0):
            intN = int(n)
            n[...] = intN*rho[intN]
#     print(nRhoValues)
    
    #integral 
    cValues = np.zeros(mValues.shape)
#     for row in range(0, mValues.shape[0]):
    for column in range(0, mValues.shape[0]):
        if(mValues[column] != 0):
            mSlice = mValues[column:]
            nRhoSlice = nRhoValues[column:]
#             print(mSlice)
#             print(nRhoSlice)
            cValues[column] = integrate.simps(nRhoSlice, mSlice)
    #mask 0s
    mValues = np.ma.masked_less_equal(mValues, 0)
    cValues = np.ma.masked_less(cValues, 0.1)
    
    #return tuple
    return (mValues, cValues)
    
print('Done: ' + time.ctime(time.time()) + " by: " + os.environ['COMPUTERNAME'])
```

# Variables for Mutations #

```python
isDebug = False
isNonStochasticSC = False #if the SCs should be decided stochastically
isDiff = True #if N_s != N_p != N_d
isRateHandSet = True #if r_p and r_d are entered by hand and Nph and Ndh are calculated from that
isDirectPControl = False #whether or not rates of inflow of SCs will be directly modified by amount of Ps
isDirectDControl = False #whether or not rates of inflow of Ps will be directly modified by amount of Ds

pControlTrigger = 1000 #number of Ps that is wanted for equilibirm
dControlTrigger = 3000 #number of Ps that is wanted for equilibirm
pControlBias = 0.2 #selective coefficient added or subtracted from SC's when P's are below or above trigger
dControlBias = 0.1 #selective coefficient added or subtracted from SC's when P's are below or above trigger

T = 1000 #total amount of generations
I = 1000 #total amount of iterations
t = 0 #current time in generations
i = 0 #current iterations
targetT = 300 #time we want to take distrubtion of how many cells of each type there are
#lists for values of # of cells at targetT
tNsList = []
tNpList = []
tNdList = []

# Ns = 1 #number of stem cells to start with
# Nsh = 1000 #homeostatic number of SCs
Np = 0 #number of progenitor cells to start with
Nd = 0 #number of differentiated cells to start with
ss = 0.01 #selective coefficient for SCs
sp = ss #selective coefficient for Ps
rs = 1 #10**(-1) #rate of SC division per time step

# if(not isRateHandSet): #if the rates will be calculated from homeostatic sizes
#     Nph = 2000 #homeostatic number of Ps
#     Ndh = 5000 #homeostatic number of Ds
#     rp = (-3*rs)/(2*sp) if not isDiff else (-3*Nsh*rs)/(2*Nph*sp) #rate of P division
#     rd = rs + rp if not isDiff else ((-3*Nsh*rs)/(2*Ndh*sp))*(1 - (2*sp)/3) #rate of D death
#     print('rp: %.4f' % rp)
#     print('rd: %.4f' % rd)
# else: #if rates are hand set
#     rp = 0.384
#     rd = 0.041
#     Nph = (-3*Nsh*rs)/(2*rp*sp)
#     Ndh = ((-3*Nsh*rs)/(2*rd*sp))*(1 - (2*sp)/3)

rp = 1
rd = 1
psDivide = 0 if isNonStochasticSC else p*rs*(1 + ss) #baseline prob. that SC will divide into 2 new SCs
psSurvive = 1*rs if isNonStochasticSC else rs*(1 - 2*p) #baseline prob. that SC will divide into 1 SC and 1 P
psDiff = 0 if isNonStochasticSC else p*rs*(1 - ss) #baseline prob. that SC will create no more SCs, dividing into 2 Ps
ppDivide = p*rp*(1 + sp) #baseline prob. that P will divide into 2 new Ps
ppSurvive = rp*(1 - 2*p) #baseline prob. that P will create only 1 more P, dividing into SC and progenitor
ppDiff = p*rp*(1 - sp) #baseline prob. that P will create no more Ps, dividing into 2 Ps
pdDie = rd #baseline prob. that D will die

mutationsNum = 1 #number of mutations
kMutants = 1 #allow cells with up to this many mutations
#create array for mutations w/ each row having mu and s boosts for each cell type:
#row: [id, mus, mup, mud, ss, sp]
mutationsArray = np.zeros((mutationsNum, 6), dtype='float32') 
mutationsArray[:, 0] = np.arange(1, mutationsNum+1) #give mutations a unique num in the 1st column
mutationsArray[:, 1] = (10**(-1))#*np.random.uniform(-1, 1, size=(mutationsNum, 3)) #give mutations random mu, ss, sp, sd

#combination includes terms that are up to and including kMutants
#num of unique cells, 1 for each combination, plus 1 w/ no mutations
uniqueCellNumArray = np.array([misc.comb(mutationsNum, k) for k in range(1, kMutants + 1)], dtype='uint32')
uniqueCellNum = uniqueCellNumArray.sum() + 1 
#row: [comb0, comb1, comb2, ... , comb(kMutants-1), totalss, totalsp, totalsd]
uniqueCellLegend = np.zeros((uniqueCellNum, kMutants + 2), dtype='float32') #each cell contains 0.0
#write combination terms into each row
index = 1 #1st row (row[0]) is wild type
for k in range(1, kMutants + 1): #for each possible # of combinations
    #input list of combination numbers going from last row index used to allow for new amount of rows to be added
    #use only the columns for number of elements of the combination there will be (this is the variable k)
    uniqueCellLegend[index:(index + uniqueCellNumArray[k-1]), 0:k] = (
        [list(i) for i in list(it.combinations(mutationsArray[:, 0], k))])
    index += uniqueCellNumArray[k-1] #increment index to reflect next starting point
    
# print(uniqueCellLegend[0:(uniqueCellNum-1), 0:kMutants])
#sum ss, sp, sd of each mutation to make totalss, totalsp, totalsd

rowIndex = 0
for row in uniqueCellLegend: #loop through each row
    totalss = 0
    totalsp = 0
    for element in row: #loop through each element
        if(element == 0): #put into uniqueCellLegend and stop when an element is 0
            uniqueCellLegend[rowIndex, kMutants] = totalss
            uniqueCellLegend[rowIndex, kMutants + 1] = totalsp
            break
        else: #increment totals by values of ss, sp, sd in mutationsArray
            totalss += mutationsArray[int(element - 1), 4]
            totalsp += mutationsArray[int(element - 1), 5]
    rowIndex += 1
#     print(row)

attributeNum = 4 #number of attributes for each set of unique cells: [N, pDivide, pSurvive, pDiff/pDie]

#3-D state cube at [i, t]
#rows for each cell type (SC, P, D), cols for binning each unique type of cell, and array for attributes
stateCube = stateCubeSetup(np.zeros((3, uniqueCellNum, attributeNum), dtype='float32'))
#after each update, state cube will be stored in a file at its appropriate [i, t] such that the full data will 5-D
#attributes etc are added w/ func stateCubeSetup

# print(rd)
print('isDebug:', isDebug)
print('isDirectPControl:', isDirectPControl)
print('isDirectDControl:', isDirectDControl)
print('T:', T, "I:", I)
print(stateCube)
print('mutationsArray:', mutationsArray)
print(uniqueCellLegend)

print('Done: ' + time.ctime(time.time()) + " by: " + os.environ['COMPUTERNAME'])
```

```python
print('Start: ' + time.ctime(time.time()) + " by: " + os.environ['COMPUTERNAME'])

#FIXED - if cellNum for Ps = 1 and 2 diff then cellNum for Ps couls become < 0

#save to file after each t or each i?
#save after each i b/c plotter can access entire i at a time, plot that line, then go to next i

# tNsList = []
# tNpList = []
# tNdList = []

#creates path name 
filePath = 'mutationsNPYs\\' + time.strftime('%m-%d-%y') + '_selection1'
if not os.path.exists(filePath):
    os.makedirs(filePath)
else:
    print('Warning: Writing over previous file!')
    time.sleep(10000)

if(isDebug): 
    debugFile = open(filePath + '\debugOutput.txt', 'w')

#create info file
infoFileName = filePath + '\info.npy'
info = np.array([T, I, rs, rp, rd, sp, isDirectPControl, pControlTrigger, pControlBias, isDirectDControl, dControlBias, dControlTrigger])
np.save(infoFileName, info)

for i in range(0, I):
    #array of all state cubes for this lineage [t, stateCube]
    lineageStateCubes = np.zeros((T, 3, uniqueCellNum, attributeNum), dtype='float32')
    stateCube = stateCubeSetup(np.zeros((3, uniqueCellNum, attributeNum), dtype='float32'))
    lineageStateCubes[0] = stateCube #add stateCube to time 0
    
    fileName = filePath + '\lineage%d.npy' % i #create file
    
    for t in range (1, T):
        if isDebug: debugFile.write('Time: ' + str(t) + '\n')
        #create a new state cube to add at this time
        newStateCube = stateCubeSetup(np.zeros((3, uniqueCellNum, attributeNum), dtype='float32'))
        newStateCube[0, 0, 0] = 0 #take out the 1 WT SC that stateCubeSetup adds
        newStateCube[0, 1, 0] = 0
        
#         if isDebug: debugFile.write('New State Cube:', newStateCube)
        #find genotypes w/ != 0 cells in them, put order pair in each row
        totalSCs = np.sum(stateCube[0, :, 0])
        totalPs = np.sum(stateCube[1, :, 0])
        totalDs = np.sum(stateCube[2, :, 0])
        nonZeroIndices = np.array(list(zip(*np.nonzero(stateCube[:, :, 0]))))
        if isDebug: debugFile.write('Non-Zero Indicies: ' + str(nonZeroIndices) + '\n')
        for orderedPair in nonZeroIndices: #ordered pair of cell type (SC, P, D) and genotype
            if isDebug: debugFile.write('Ordered Pair: ' + str(tuple(orderedPair)) + '\n')
#             print(stateCube[tuple(orderedPair)][0])
#             print(stateCube[0,0,0])
            cellType = orderedPair[0] #SC = 0, P = 1, D = 2
            cellNum = stateCube[tuple(orderedPair)][0]
            pDiv = stateCube[tuple(orderedPair)][1]
            pSurv = stateCube[tuple(orderedPair)][2]
            pDiff = stateCube[tuple(orderedPair)][3]
            if isDebug: debugFile.write('Cell Num: %d, pDiv: %.2f, pSurv: %.2f, pDiff: %.2f \n' % (cellNum, pDiv, pSurv, pDiff))
            currentMutations = uniqueCellLegend[orderedPair[1], :kMutants] #all the mutations that cells in this compartment have
            isWT = False
            mutationsCanAcquire = np.copy(mutationsArray[:, 0])
#             print(mutationsArray)
#             print('currentMutations: ', currentMutations)
#             print('mutationsArray1D 1: ', mutationsArray1D)
            if((currentMutations==0).all()): #if this is a WT cell
                isWT = True
            else:
                for mut in currentMutations:
                    if((mutationsCanAcquire==mut).any() and mut != 0): 
                        mutationsCanAcquire[int(mut)-1] = 0

            if isDebug: debugFile.write('Mutations Can Acquire: ' + str(mutationsCanAcquire) + '\n')
#             mutationsCanAcquire = [mut[0] for mut in mutationsArray if (not (mut==currentMutations).any())] #mutations that cell can acquire are ones it doesn't already have
#             print(mutationsArray)
#             print(mutationsCanAcquire[0])
            
            while (True): #do while
                mutDivNum = np.arange(mutationsNum+1)
                mutSurvNum = np.arange(mutationsNum+1)
                mutDiffNum = np.arange(mutationsNum+1) 

                if(cellType != 2): #only run these Poissons if SC or P since Ds only diff
                    if(isDirectPControl and (cellType == 0)): #only change probabilites for SCs
                        if(totalPs > pControlTrigger): #if there are too many Ps, bias to symmetric division into 2 SCs
                            pDiv += rs*pControlBias
                            pDiff -= rs*pControlBias
                        elif(totalPs < pControlTrigger): #if there are too few Ps, bias to symmetric division into 2 Ps
                            pDiff += rs*pControlBias
                            pDiv -= rs*pControlBias
                        if isDebug: debugFile.write('new pDiv: ' + str(pDiv) + ', new pDiff: ' + str(pDiff) + '\n')
                    if(isDirectDControl and (cellType == 1)):
                        if(totalDs > dControlTrigger): #if there are too many Ds, bias to symmetric division into 2 Ps
                            pDiv += rp*dControlBias
                            pDiff -= rp*dControlBias
                        elif(totalDs < dControlTrigger): #if there are too few Ds, bias to symmetric division into 2 Ds
                            pDiff += rp*dControlBias
                            pDiv -= rp*dControlBias
                        if isDebug: debugFile.write('new pDiv: ' + str(pDiv) + ', new pDiff: ' + str(pDiff) + '\n')
                        
                    #prob of no mutations = pDiv*(1 - sum(m=1, 10, mu))
#                     if(cellNum < 0):
#                         print('Cell Num:', cellNum, tuple(orderedPair), t)
                    #use cellType+1 to get the correct mu value
                    if isDebug: 
                        debugFile.write('Part of lambda from muts: ' + str(1 - sum([mutationsArray[a-1, cellType+1] for a in mutationsCanAcquire if (a != 0)])) + '\n')
                        debugFile.write('Part of lambda, cellNum: ' + str(cellNum) + '\n')
                        debugFile.write('Lambda: ' + str(pDiv*cellNum*(1 - sum([mutationsArray[a-1, cellType+1] for a in mutationsCanAcquire if (a != 0)]))) + '\n')
                    mutDivNum[0] = np.random.poisson(pDiv*cellNum*(1 - sum([mutationsArray[a-1, cellType+1] for a in mutationsCanAcquire if (a != 0)])))   
                    #probablity of mutation m = mu*pDiv
                    #pDivMut[m] = mutationsArray[m - 1, cellType+1]*stateCube[*orderPair, 1]
                    for index, divNum in np.ndenumerate(mutDivNum[1:]): #number that acquire each mutation
                        if ((mutationsCanAcquire==divNum).any()): #only run poisson if doesn't already have this mutation
                            mutDivNum[index[0]+1] = np.random.poisson(mutationsArray[divNum-1, cellType+1]*pDiv*cellNum)
                        else:
                             mutDivNum[index[0]+1] = 0

#                     if(cellNum < 0):
#                         print('Cell Num:', cellNum, tuple(orderedPair), t)
                    mutSurvNum[0] = np.random.poisson(pSurv*cellNum*(1 - sum([mutationsArray[a-1, cellType+1] for a in mutationsCanAcquire if (a != 0)])))

                    for index, divNum in np.ndenumerate(mutSurvNum[1:]):
#                         print(divNum)
#                         print(index[0]+1)
                        if ((mutationsCanAcquire==divNum).any()): #only run poisson if doesn't already have this mutation
                            mutSurvNum[index[0]+1] = np.random.poisson(mutationsArray[divNum-1, cellType+1]*pSurv*cellNum)
                        else:
                            mutSurvNum[index[0]+1] = 0
                else: #if D, set div and surv to 0
                    if(cellNum < 0): #clipping for now
                        cellNum = 0
                    mutDivNum[:] = [0]*len(mutDivNum)
                    mutSurvNum[:] = [0]*len(mutSurvNum)

#                 if(cellNum < 0):
#                     print('Cell Num:', cellNum, tuple(orderedPair), t)
                #run these for any cell type
                mutDiffNum[0] = np.random.poisson(pDiff*cellNum*(1 - sum([mutationsArray[a-1, cellType+1] for a in mutationsCanAcquire if (a != 0)])))

                for index, divNum in np.ndenumerate(mutDiffNum[1:]):
                    if ((mutationsCanAcquire==divNum).any()): #only run poisson if doesn't already have this mutation
                        mutDiffNum[index[0]+1] = np.random.poisson(mutationsArray[divNum-1, cellType+1]*pDiff*cellNum)
                    else:
                        mutDiffNum[index[0]+1] = 0

                if(cellNum + sum(mutDivNum) - sum(mutDiffNum) >= 0 or cellType == 2): #check that new # of cells is positive
                    if isDebug:
                        debugFile.write('cellNum: '+ str(cellNum) + '\n')
                        #adds * around number if != 0
                        debugFile.write('mutDivNum Same: ' + str(mutDivNum[0]) + '\n') if (mutDivNum[0] == 0) else debugFile.write('mutDivNum Same: *' + str(mutDivNum[0]) + '*' + '\n')
                        debugFile.write('mutDivNum Mutation 1: ' + str(mutDivNum[1]) + '\n') if (mutDivNum[1] == 0) else debugFile.write('mutDivNum Mutation 1: *' + str(mutDivNum[1]) + '*' + '\n')
                        debugFile.write('mutSurvNum Same: ' + str(mutSurvNum[0]) + '\n') if (mutSurvNum[0] == 0) else debugFile.write('mutSurvNum Same: *' + str(mutSurvNum[0]) + '*' + '\n')
                        debugFile.write('mutSurvNum Mutation 1: ' + str(mutSurvNum[1]) + '\n') if (mutSurvNum[1] == 0) else debugFile.write('mutSurvNum Mutation 1: *' + str(mutSurvNum[1]) + '*' + '\n')
                        debugFile.write('mutDiffNum Same: ' + str(mutDiffNum[0]) + '\n') if (mutDiffNum[0] == 0) else debugFile.write('mutDiffNum Same: *' + str(mutDiffNum[0]) + '*' + '\n')
                        debugFile.write('mutDiffNum Mutation 1: ' + str(mutDiffNum[1]) + '\n') if (mutDiffNum[1] == 0) else debugFile.write('mutDiffNum Mutation 1: *' + str(mutDiffNum[1]) + '*' + '\n')
                    break

    #                 if((sum(mutDivNum) + sum(mutSurvNum) + sum(mutDiffNum)) <= cellNum): #check to make sure didn't generate more divisions than possible
    #             print('mutDivNum: ', sum(mutDivNum))
    #             print('mutSurvNum: ', sum(mutSurvNum))
    #             print('mutDiffNum: ', sum(mutDiffNum))
    #                     break
        
            #put into compartments by genotype
            doNothing = cellNum - (sum(mutDivNum) + sum(mutSurvNum) + sum(mutDiffNum)) #cells that didn't do anything
#             doNothing += mutDivNum[0] #add 2*cells dividing w/ 0 mutations
            if isDebug: debugFile.write('Cells Staying in Same Compartment and not diving:' + str(doNothing) + '\n')
            newStateCube[tuple(orderedPair)][0] += doNothing #add back to same compartment
            
            if(cellType != 2): #not D
                genotypes = [[*currentMutations, acquireable] for acquireable in mutationsCanAcquire if (acquireable != 0)] #2D list of lists of genotypes
                genotypes.append([*currentMutations, 0.]) #add that it can 
#                 print('Genotypes to enter in newStateCube:', genotypes)
                for genotype in genotypes:
                    newMutation = int(genotype[len(currentMutations)]) #the mutation acquired
#                     print(newMutation)
                    genotypeCompSame = 0 #number going into compartment of new genotype, same cell level (SC or P)
                    genotypeCompDown = 0 #number going into compartment of new genotype, 1 cell level down (SC->P or P->D)
                    genotypeCompSame += 2*mutDivNum[newMutation] + mutSurvNum[newMutation]
                    genotypeCompDown += mutSurvNum[newMutation] + 2*mutDiffNum[newMutation]
                    if isDebug: debugFile.write('genotypeCompSame ' + str(genotypeCompSame) + ' genotypeCompDown ' + str(genotypeCompDown) + '\n')
                    index = 0 #the index in the state cube of the genotype
                    if(genotypeCompSame != 0 or genotypeCompDown != 0 and newMutation != 0): #if there's something to add and is getting a new mutation
                        #find the index of this genotype
                        genotypeNoZeros = sorted([g for g in genotype if g != 0]) #take out 0s and leave only mutations
                        cellLegendSlice = uniqueCellLegend[:, :len(genotypeNoZeros)] #slice of all genotypes of only length of
                        for geno in cellLegendSlice:
                            if(list(geno) == genotypeNoZeros):
                                break
                            else:
                                index += 1
                                
                    #add to state cube
#                     print(index)
                    newStateCube[cellType, orderedPair[1] if index == 0 else index, 0] += genotypeCompSame
                    newStateCube[cellType+1, orderedPair[1] if index == 0 else index, 0] += genotypeCompDown
                    
#         if(t == targetT): #if this is time to measure # of cells
#             tNsList.append(np.sum(newStateCube[0, :, 0])) 
#             tNpList.append(np.sum(newStateCube[1, :, 0]))
#             tNdList.append(np.sum(newStateCube[2, :, 0]))
            
#         if(cellNum + sum(mutDivNum) - sum(mutDiffNum) < 0):
#             print('cellNum after: ', newStateCube[tuple(orderedPair)][0])
#             if(newStateCube[tuple(orderedPair)][0] >= 0):
#                 print('State Cube:\n', newStateCube, '\n')
        
        newStateCube[newStateCube < 0] = 0 #clip all negative values
        
        lineageStateCubes[t, :, :, :] = np.copy(newStateCube)
        stateCube = np.copy(newStateCube)
#         if(t <= 50):
        if isDebug: 
            debugFile.write('Total SCs (previous t): ' + str(totalSCs) + '\n')
            debugFile.write('Total Ps (previous t): ' + str(totalPs) + '\n')
            debugFile.write('Total Ds (previous t): ' + str(totalDs) + '\n')
            debugFile.write('State Cube:\n')
            debugFile.write(str(newStateCube))
            debugFile.write('\n\n')
        nonZeroIndicies = None
        orderedPair = None
                
    #save iteration
#     print('File Name:', fileName)
#     print('Entire Lineage:', lineageStateCubes)
    np.save(fileName, lineageStateCubes)

#save distribution arrays
# np.save(filePath + '\tNsList.npy', tNsList)
# np.save(filePath + '\tNpList.npy', tNpList)
# np.save(filePath + '\tNdList.npy', tNdList)

if(isDebug):
    debugFile.close()

print('Done: ' + time.ctime(time.time()) + " by: " + os.environ['COMPUTERNAME'])
```

# Expectations and Math 3 #
## Neutral Feeding Non-Neutral (Ps feeding Ds) ##
Defs:
* $n_p(t)$ proenitor cells w/ a certain mutation neutrally feed $n_d(t)$ differentiated cells w/ that mutations
* $s_p$ is the negligble selective bias for progenitor cells
* $r_p$ is the rate of progenitor division such that $n_p$ changes, through neutral dynamics only, by $\frac{1}{n_p}$ in $\frac{1}{n_p}$ time
* $r_d$ is the rate of differentiated cell death

$\frac{dn_d}{dt}=$ number entering from neutral feeding - number going out from death <br>
$=\frac{2}{3}n_p r_p+\frac{1}{3}n_pr_p-n_dr_d$ <br>
$\approx\frac{2}{3}n_pr_p+\frac{1}{3}n_pr_p-n_dr_d$ <br>
$\frac{dn_d}{dt}\approx n_pr_p-n_dr_d$ <br>
<br>
Consider $\frac{dn_d}{dt}=0$: <br>
$n_pr_p=n_dr_d$ <br>
$n_d(t)=\frac{n_p(t)r_p}{r_d}$ <br>
**_Stable fixed point: $F_d(t)=\frac{r_p}{r_d}\cdot n_p(t)$ depends only on $n_p(t)$_** <br>
<br>
*Drop 1st term, investigate 2nd term:* <br>
Let $\frac{dn_d}{dt}=-n_dr_d$ <br>
$\int{-\frac{dn_d}{n_dr_d}}=\int{dt}$ <br>
$-\frac{\ln{|n_d|}}{r_d}=t+C$ <br>
$\ln{n_d}=-r_dt+C$ <br>
$n_d=Ce^{-r_dt}$ <br>
$n_d(t)=n_0e^{-r_dt}$ <br>
**_Exponential decay over $\frac{1}{r_d}$ timescale (to change by $n_p(t)\cdot r_p$)_** <br>
<br>
*Drop 2nd term, investigate 1st term:* <br>
$\frac{dn_p}{dt}=n_pr_p$ <br>
$n_d(t)=n_pr_pt$ **linear increase** over ?? timescale <br>
Timescale: <br>
If $n_d(t)\gg F_d(t)\Rightarrow$ exponential decay over $\frac{1}{r_d}$ timescale <br>
If $n_d(t)\ll F_d(t)\Rightarrow$ linear increase over $\frac{1}{r_d}$ timescale to change by a significant amount (ie. $n_p(t)\cdot r_p$) as well <br>
**_Same timescale of increase and decrease back to $F_d(t)$_** <br>
<br>
**Chasing Argument** <br>
$F_d(t)$ is changing in time while $n_d(t)$ is trying to catch up to or "chase" it. Observations: <br>
* if $r_d\gg r_p \Rightarrow n_d(t)$ will closely hug $F_d(t)$ because $\frac{1}{r_d}$ is of the same order on which $F_d(t)$ changes, which is also $\frac{1}{r_d}$.
    * $\Rightarrow n_d(t) \propto n_p(t)$
* if $r_d \ll r_p$: compare the times in which $n_d(t)$ tries to reach $F_d(t)$ while $F_d(t)$ changes
    * $n_d(t_0)$
        * Takes time $\frac{1}{r_d}$ to move to $F_d(t_0)$
        * This move is a distance of $d=F_d(t_0)-n_d(t_0)$ such that $n_d(t_0+\frac{1}{r_d}=F_d(t_0)$
    * At the same time, $F_d(t_0)$
        * Takes time $\frac{1}{r_d}$
        * Since $F_d(t)$ only depends on the neutral dynamics of $n_p(t)$, over a time of $\frac{1}{r_d}$, $n_p(t)$ will change by $\frac{r_p}{r_d}$ (must multiply by $r_p$ because of adjusment for rates of division)
        * $\Rightarrow d=\frac{r_p}{r_d^2}$ 
        <br>
        
This **_$\frac{r_p}{r_d^2}$ is the flucation_** that will be seen from our expected number of differentiated cells (which we approximate with $F_d(t)$). <br>
**_$\epsilon \sim \frac{r_p}{r_d^2}$_**

```python
print('Start: ' + time.ctime(time.time()) + " by: " + os.environ['COMPUTERNAME'])

#FIXED - if cellNum for Ps = 1 and 2 diff then cellNum for Ps couls become < 0

#save to file after each t or each i?
#save after each i b/c plotter can access entire i at a time, plot that line, then go to next i

# tNsList = []
# tNpList = []
# tNdList = []

#creates path name 
filePath = 'mutationsNPYs\\' + time.strftime('%m-%d-%y') + '_nfnPerfectScTriangle9'
if not os.path.exists(filePath):
    os.makedirs(filePath)
else:
    print('Warning: Writing over previous file!')
    time.sleep(10000)

if(isDebug): 
    debugFile = open(filePath + '\debugOutput.txt', 'w')

#create info file
infoFileName = filePath + '\info.npy'
info = np.array([T, I, rs, rp, rd, sp, te, maxNste])
np.save(infoFileName, info)

for i in range(0, I):
    #array of all state cubes for this lineage [t, stateCube]
    lineageStateCubes = np.zeros((T, 3, uniqueCellNum, attributeNum), dtype='float32')
    stateCube = stateCubeSetup(np.zeros((3, uniqueCellNum, attributeNum), dtype='float32'))
    lineageStateCubes[0] = stateCube #add stateCube to time 0
    
    fileName = filePath + '\lineage%d.npy' % i #create file
    
    for t in range (1, T):
        if isDebug: debugFile.write('Time: ' + str(t) + '\n')
        #create a new state cube to add at this time
        newStateCube = stateCubeSetup(np.zeros((3, uniqueCellNum, attributeNum), dtype='float32'))
        newStateCube[0, 0, 0] = 0 #take it out the 1 WT SC that stateCubeSetup adds
        newStateCube[0, 1, 0] = 0
        
#         if isDebug: debugFile.write('New State Cube:', newStateCube)
        #find genotypes w/ != 0 cells in them, put order pair in each row
        nonZeroIndices = np.array(list(zip(*np.nonzero(stateCube[:, :, 0]))))
        if isDebug: debugFile.write('Non-Zero Indicies: ' + str(nonZeroIndices) + '\n')
        for orderedPair in nonZeroIndices: #ordered pair of cell type (SC, P, D) and genotype
            if isDebug: debugFile.write('Ordered Pair: ' + str(tuple(orderedPair)) + '\n')
#             print(stateCube[tuple(orderedPair)][0])
#             print(stateCube[0,0,0])
            cellType = orderedPair[0] #SC = 0, P = 1, D = 2
            cellNum = stateCube[tuple(orderedPair)][0]
            pDiv = stateCube[tuple(orderedPair)][1]
            pSurv = stateCube[tuple(orderedPair)][2]
            pDiff = stateCube[tuple(orderedPair)][3]
            if isDebug: debugFile.write('Cell Num: %d, pDiv: %.2f, pSurv: %.2f, pDiff: %.2f \n' % (cellNum, pDiv, pSurv, pDiff))
            currentMutations = uniqueCellLegend[orderedPair[1], :kMutants] #all the mutations that cells in this compartment have
            isWT = False
            mutationsCanAcquire = np.copy(mutationsArray[:, 0])
#             print(mutationsArray)
#             print('currentMutations: ', currentMutations)
#             print('mutationsArray1D 1: ', mutationsArray1D)
            if((currentMutations==0).all()): #if this is a WT cell
                isWT = True
            else:
                for mut in currentMutations:
                    if((mutationsCanAcquire==mut).any() and mut != 0): 
                        mutationsCanAcquire[mut-1] = 0

            if isDebug: debugFile.write('Mutations Can Acquire: ' + str(mutationsCanAcquire) + '\n')
#             mutationsCanAcquire = [mut[0] for mut in mutationsArray if (not (mut==currentMutations).any())] #mutations that cell can acquire are ones it doesn't already have
#             print(mutationsArray)
#             print(mutationsCanAcquire[0])
            
            while (True): #do while
                mutDivNum = np.arange(mutationsNum+1)
                mutSurvNum = np.arange(mutationsNum+1)
                mutDiffNum = np.arange(mutationsNum+1) 

                if(cellType == 1): #only run these Poissons if P since Ds only diff and SC need to make triangle
                    #prob of no mutations = pDiv*(1 - sum(m=1, 10, mu))
#                     if(cellNum < 0):
#                         print('Cell Num:', cellNum, tuple(orderedPair), t)
                    #use cellType+1 to get the correct mu value
                    mutDivNum[0] = np.random.poisson(pDiv*cellNum*(1 - sum([mutationsArray[a-1, cellType+1] for a in mutationsCanAcquire if (a != 0)])))
                    #probablity of mutation m = mu*pDiv
                    #pDivMut[m] = mutationsArray[m - 1, cellType+1]*stateCube[*orderPair, 1]
                    for index, divNum in np.ndenumerate(mutDivNum[1:]): #number that acquire each mutation
                        if ((mutationsCanAcquire==divNum).any()): #only run poisson if doesn't already have this mutation
                            mutDivNum[index[0]+1] = np.random.poisson(mutationsArray[divNum-1, cellType+1]*pDiv*cellNum)
                        else:
                             mutDivNum[index[0]+1] = 0

#                     if(cellNum < 0):
#                         print('Cell Num:', cellNum, tuple(orderedPair), t)
                    mutSurvNum[0] = np.random.poisson(pSurv*cellNum*(1 - sum([mutationsArray[a-1, cellType+1] for a in mutationsCanAcquire if (a != 0)])))

                    for index, divNum in np.ndenumerate(mutSurvNum[1:]):
#                         print(divNum)
#                         print(index[0]+1)
                        if ((mutationsCanAcquire==divNum).any()): #only run poisson if doesn't already have this mutation
                            mutSurvNum[index[0]+1] = np.random.poisson(mutationsArray[divNum-1, cellType+1]*pSurv*cellNum)
                        else:
                            mutSurvNum[index[0]+1] = 0
                        
                elif(cellType == 0): #run SCs
                    if(isPerfectTriangle):
                        if(t < te/2): #if SCs should be rising
                            mutDivNum[0] = ma.floor((te/2)/(maxNste)) #slope
                            mutDiffNum[0] = 0
                        else: #SCs should be decreasing
                            mutDivNum[0] = 0
                            mutDiffNum[0] = ma.floor((te/2)/(maxNste)) #negative slope
                            
                        for index, divNum in np.ndenumerate(mutDivNum[1:]): #number that acquire each mutation (0 for triangle test)
                            if ((mutationsCanAcquire==divNum).any()): #only run poisson if doesn't already have this mutation
                                mutDivNum[index[0]+1] = 0
                            else:
                                 mutDivNum[index[0]+1] = 0

                        mutSurvNum[0] = cellNum - mutDivNum[0] - mutDiffNum[0] #the rest will surv

                        for index, divNum in np.ndenumerate(mutSurvNum[1:]):
    #                         print(divNum)
    #                         print(index[0]+1)
                            if ((mutationsCanAcquire==divNum).any()): #only run poisson if doesn't already have this mutation
                                mutSurvNum[index[0]+1] = 0
                            else:
                                mutSurvNum[index[0]+1] = 0
                            
                    else: #regular, no perfect triangle
                        #prob of no mutations = pDiv*(1 - sum(m=1, 10, mu))
#                     if(cellNum < 0):
#                         print('Cell Num:', cellNum, tuple(orderedPair), t)
                        #use cellType+1 to get the correct mu value
                        mutDivNum[0] = np.random.poisson(pDiv*cellNum*(1 - sum([mutationsArray[a-1, cellType+1] for a in mutationsCanAcquire if (a != 0)])))
                        #probablity of mutation m = mu*pDiv
                        #pDivMut[m] = mutationsArray[m - 1, cellType+1]*stateCube[*orderPair, 1]
                        for index, divNum in np.ndenumerate(mutDivNum[1:]): #number that acquire each mutation
                            if ((mutationsCanAcquire==divNum).any()): #only run poisson if doesn't already have this mutation
                                mutDivNum[index[0]+1] = np.random.poisson(mutationsArray[divNum-1, cellType+1]*pDiv*cellNum)
                            else:
                                 mutDivNum[index[0]+1] = 0

    #                     if(cellNum < 0):
    #                         print('Cell Num:', cellNum, tuple(orderedPair), t)
                        mutSurvNum[0] = np.random.poisson(pSurv*cellNum*(1 - sum([mutationsArray[a-1, cellType+1] for a in mutationsCanAcquire if (a != 0)])))

                        for index, divNum in np.ndenumerate(mutSurvNum[1:]):
    #                         print(divNum)
    #                         print(index[0]+1)
                            if ((mutationsCanAcquire==divNum).any()): #only run poisson if doesn't already have this mutation
                                mutSurvNum[index[0]+1] = np.random.poisson(mutationsArray[divNum-1, cellType+1]*pSurv*cellNum)
                            else:
                                mutSurvNum[index[0]+1] = 0
                    
                else: #if D, set div and surv to 0
                    if(cellNum < 0): #clipping for now
                        cellNum = 0
                    mutDivNum[:] = [0]*len(mutDivNum)
                    mutSurvNum[:] = [0]*len(mutSurvNum)

#                 if(cellNum < 0):
#                     print('Cell Num:', cellNum, tuple(orderedPair), t)
                #run these for any cell type
                if(not (cellType == 0 and isPerfectTriangle)): #only run if not SC and not perfect triangle so as not to override the muDiffNum[0] set above
                    mutDiffNum[0] = np.random.poisson(pDiff*cellNum*(1 - sum([mutationsArray[a-1, cellType+1] for a in mutationsCanAcquire if (a != 0)])))

                for index, divNum in np.ndenumerate(mutDiffNum[1:]):
                    if ((mutationsCanAcquire==divNum).any()): #only run poisson if doesn't already have this mutation
                        mutDiffNum[index[0]+1] = np.random.poisson(mutationsArray[divNum-1, cellType+1]*pDiff*cellNum)
                    else:
                        mutDiffNum[index[0]+1] = 0

                if(cellNum + sum(mutDivNum) - sum(mutDiffNum) >= 0 or cellType == 2): #check that new # of cells is positive
                    if isDebug:
                        debugFile.write('cellNum: '+ str(cellNum) + '\n')
                        #adds * around number if != 0
                        debugFile.write('mutDivNum Same: ' + str(mutDivNum[0]) + '\n') if (mutDivNum[0] == 0) else debugFile.write('mutDivNum Same: *' + str(mutDivNum[0]) + '*' + '\n')
                        debugFile.write('mutDivNum Mutation 1: ' + str(mutDivNum[1]) + '\n') if (mutDivNum[1] == 0) else debugFile.write('mutDivNum Mutation 1: *' + str(mutDivNum[1]) + '*' + '\n')
                        debugFile.write('mutSurvNum Same: ' + str(mutSurvNum[0]) + '\n') if (mutSurvNum[0] == 0) else debugFile.write('mutSurvNum Same: *' + str(mutSurvNum[0]) + '*' + '\n')
                        debugFile.write('mutSurvNum Mutation 1: ' + str(mutSurvNum[1]) + '\n') if (mutSurvNum[1] == 0) else debugFile.write('mutSurvNum Mutation 1: *' + str(mutSurvNum[1]) + '*' + '\n')
                        debugFile.write('mutDiffNum Same: ' + str(mutDiffNum[0]) + '\n') if (mutDiffNum[0] == 0) else debugFile.write('mutDiffNum Same: *' + str(mutDiffNum[0]) + '*' + '\n')
                        debugFile.write('mutDiffNum Mutation 1: ' + str(mutDiffNum[1]) + '\n') if (mutDiffNum[1] == 0) else debugFile.write('mutDiffNum Mutation 1: *' + str(mutDiffNum[1]) + '*' + '\n')
                    break

    #                 if((sum(mutDivNum) + sum(mutSurvNum) + sum(mutDiffNum)) <= cellNum): #check to make sure didn't generate more divisions than possible
    #             print('mutDivNum: ', sum(mutDivNum))
    #             print('mutSurvNum: ', sum(mutSurvNum))
    #             print('mutDiffNum: ', sum(mutDiffNum))
    #                     break
        
            #put into compartments by genotype
            doNothing = cellNum - (sum(mutDivNum) + sum(mutSurvNum) + sum(mutDiffNum)) #cells that didn't do anything
#             doNothing += mutDivNum[0] #add 2*cells dividing w/ 0 mutations
            if isDebug: debugFile.write('Cells Staying in Same Compartment and not diving:' + str(doNothing) + '\n')
            newStateCube[tuple(orderedPair)][0] += doNothing #add back to same compartment
            
            if(cellType != 2): #not D
                genotypes = [[*currentMutations, acquireable] for acquireable in mutationsCanAcquire if (acquireable != 0)] #2D list of lists of genotypes
                genotypes.append([*currentMutations, 0.]) #add that it can 
#                 print('Genotypes to enter in newStateCube:', genotypes)
                for genotype in genotypes:
                    newMutation = genotype[len(currentMutations)] #the mutation acquired
#                     print(newMutation)
                    genotypeCompSame = 0 #number going into compartment of new genotype, same cell level (SC or P)
                    genotypeCompDown = 0 #number going into compartment of new genotype, 1 cell level down (SC->P or P->D)
                    genotypeCompSame += 2*mutDivNum[newMutation] + mutSurvNum[newMutation]
                    genotypeCompDown += mutSurvNum[newMutation] + 2*mutDiffNum[newMutation]
                    if isDebug: debugFile.write('genotypeCompSame ' + str(genotypeCompSame) + ' genotypeCompDown ' + str(genotypeCompDown) + '\n')
                    index = 0 #the index in the state cube of the genotype
                    if(genotypeCompSame != 0 or genotypeCompDown != 0 and newMutation != 0): #if there's something to add and is getting a new mutation
                        #find the index of this genotype
                        genotypeNoZeros = sorted([g for g in genotype if g != 0]) #take out 0s and leave only mutations
                        cellLegendSlice = uniqueCellLegend[:, :len(genotypeNoZeros)] #slice of all genotypes of only length of
                        for geno in cellLegendSlice:
                            if(list(geno) == genotypeNoZeros):
                                break
                            else:
                                index += 1
                                
                    #add to state cube
#                     print(index)
                    newStateCube[cellType, orderedPair[1] if index == 0 else index, 0] += genotypeCompSame
                    newStateCube[cellType+1, orderedPair[1] if index == 0 else index, 0] += genotypeCompDown
                    
#         if(t == targetT): #if this is time to measure # of cells
#             tNsList.append(np.sum(newStateCube[0, :, 0])) 
#             tNpList.append(np.sum(newStateCube[1, :, 0]))
#             tNdList.append(np.sum(newStateCube[2, :, 0]))
            
#         if(cellNum + sum(mutDivNum) - sum(mutDiffNum) < 0):
#             print('cellNum after: ', newStateCube[tuple(orderedPair)][0])
#             if(newStateCube[tuple(orderedPair)][0] >= 0):
#                 print('State Cube:\n', newStateCube, '\n')
                
        lineageStateCubes[t, :, :, :] = np.copy(newStateCube)
        stateCube = np.copy(newStateCube)
#         if(t <= 50):
        if isDebug: 
            debugFile.write('State Cube:\n')
            debugFile.write(str(newStateCube))
            debugFile.write('\n\n')
        nonZeroIndicies = None
        orderedPair = None
                
    #save iteration
#     print('File Name:', fileName)
#     print('Entire Lineage:', lineageStateCubes)
    np.save(fileName, lineageStateCubes)

#save distribution arrays
# np.save(filePath + '\tNsList.npy', tNsList)
# np.save(filePath + '\tNpList.npy', tNpList)
# np.save(filePath + '\tNdList.npy', tNdList)

if(isDebug):
    debugFile.close()

print('Done: ' + time.ctime(time.time()) + " by: " + os.environ['COMPUTERNAME'])
```
